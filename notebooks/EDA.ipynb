# Distribution of Numerical Features


```python
#Visualize the distribution of numerical features to identify patterns, skewness, and potential outliers.
```

```python
numerical_columns = data.select_dtypes(include=['int', 'float'])
```

```python
# Plot histograms for numerical features
for column in numerical_columns.columns:
    
    sns.histplot(data=numerical_columns[column], kde=True)
    
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Plot box plots for numerical features
for column in numerical_columns.columns:
    
    sns.boxplot(x=numerical_columns[column])
    
    plt.xlabel(column)
    plt.show()
```

```python
# Plot density plots for numerical features
for column in numerical_columns.columns:
    
    sns.kdeplot(data=numerical_columns[column], shade=True)
    
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
```
